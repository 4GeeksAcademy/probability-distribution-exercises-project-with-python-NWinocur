## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

#### Normal Distribution
##### Defining probability functions

```python
from scipy.stats import norm

# Define all the probability functions related to this distribution

def dnorm(x, mean=0, standard_deviation=1):
    """
    Calculates the point estimate of the normal distribution
    """
    result = norm.pdf(x, loc=mean, scale=standard_deviation)

    return result

def pnorm(x, mean=0, standard_deviation=1):
    """
    Calculates the cumulative of the normal distribution
    """
    result = norm.cdf(x, loc=mean, scale=standard_deviation)

    return result

def qnorm(x, mean=0, standard_deviation=1):
    """
    Calculates the quantile function from the normal distribution
    """
    result = norm.ppf(x, loc=mean, scale=standard_deviation)

    return result

def rnorm(size, mean=0, standard_deviation=1):
    """
    Generates random variables from the normal distribution
    """
    result = norm.rvs(loc= mean, scale=standard_deviation, size=size)

    return result
```

##### Printing probability statements about normal distribution

```python
import numpy as np

np.random.seed(42)
mean_loc = 0
sd = 1

print(f"Probability that a normal distribution takes the value 1.96: {dnorm(1.96, mean=mean_loc, standard_deviation=sd)}")
print(f"Probability that a normal distribution will take a value less than 1.96: {pnorm(1.96, mean=mean_loc, standard_deviation=sd)}")
print(f"Which value of a normal distribution has 97.5 of observations under it? {qnorm(0.975, mean=mean_loc, standard_deviation=sd)}")
print(f"Generate 2000 random values from a normal distribution: {rnorm(2000, mean=mean_loc, standard_deviation=sd)}")
```

##### Plotting

```python
import matplotlib.pyplot as plt

np.random.seed(42)

sd = 1
loc = mean_loc
size = 2000
data = rnorm(size, loc, sd)

pmf = dnorm(data, loc, sd)
cdf = pnorm(data, loc, sd)
ppf = qnorm(data, loc, sd)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

#### Uniform Distribution
##### Defining probability functions

```python
from scipy.stats import uniform

# Define all the probability functions related to this distribution

def duniform(x, loc, scale):
    """
    Calculates the point estimate of the uniform distribution
    """
    result = uniform.pdf(x, loc=loc, scale=scale)

    return result

def puniform(x, loc, scale):
    """
    Calculates the cumulative of the uniform distribution
    """
    result = uniform.cdf(x, loc=loc, scale=scale)

    return result

def quniform(x, loc, scale):
    """
    Calculates the quantile function from the uniform distribution
    """
    result = uniform.ppf(x, loc=loc, scale=scale)

    return result

def runiform(size, loc=0, scale=1):
    """
    Generates random variables from the uniform distribution
    """
    result = uniform.rvs(loc=loc, scale=scale, size=size)

    return result
```

##### Printing probability statements about uniform distribution

```python
import numpy as np

np.random.seed(42)
uniform_loc = 0

print(f"Probability that a uniform distribution takes the value 1.96: {duniform(1.96, loc=uniform_loc, scale=1)}")
print(f"Probability that a uniform distribution will take a value less than 1.96: {puniform(1.96, loc=uniform_loc, scale=1)}")
print(f"Which value of a uniform distribution has 97.5 of observations under it? {quniform(0.975, loc=0, scale=1)}")
print(f"Generate 2000 random values from a uniform distribution: {runiform(2000, loc=uniform_loc, scale=1)}")
```
